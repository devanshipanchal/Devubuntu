```python
import spacy
nlp = spacy.load("en_core_web_sm")

text = "Devanshi is learning natural language processing."
doc = nlp(text)

for token in doc:
    print(token.text, "→", token.pos_)
```

1. spaCy kya hai?

spaCy = Python ka fastest, most powerful NLP library.

Iske through aap kar sakte ho:

Tokenization

POS Tagging

Lemmatization

NER (Named Entity Recognition)

Dependency Parsing

Sentence segmentation

Word vectors

Industry level tools, Rasa bhi spaCy use karta hai.

2. en_core_web_sm kya hai?

spaCy mein language models hote hain.

en_core_web_sm =

en → English

core → core pipeline

web → trained on web text

sm → small (lightweight model)

✔ Yeh aapka “English brain” hai jo spaCy ko English understand karne deta hai.
Iske bina spaCy POS tagging nahi kar sakta.

spaCy main 3 English models hote hain:


Model	Size	Speed	Accuracy

en_core_web_sm	Small	Fast	Good


en_core_web_md	Medium	Slower	Better


en_core_web_lg	Large	Slow	Best
